```python
from PIL import Image
import matplotlib.pyplot as plt
import os,cv2
import numpy as np
imgpath = r'D:\BaiduNetdiskDownload\datasets\all'
imgs = os.listdir(imgpath)
#i = 0
for img in imgs[1:2]:
    #i += 1
    # image = Image.open(os.path.join(imgpath, img))
    plt.subplot(1, 2, 1)
    image = cv2.imread(os.path.join(imgpath, img),0)
    # roi_l = image.crop((0,0,300,400))
    roi_l = image[0:400, 0:290]
    ret, thres = cv2.threshold(roi_l,100,255,cv2.THRESH_BINARY)
    contours,hierarchy = cv2.findContours(thres,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    print("number of contours:%d" % len(contours))
    #找到最大区域并填充
    area = []
    for i in range(len(contours)):
        area.append(cv2.contourArea(contours[i]))
    max_idx = np.argmax(area)
    for k in range(len(contours)):
        if k != max_idx:
            cv2.fillPoly(thres, [contours[k]], 255)
    
    # cv2.drawContours(image, contours, -1, (0, 255, 255), 2)
    # opening = cv2.morphologyEx(thres, cv2.MORPH_OPEN, np.ones((5,5),np.uint8))
    thres[thres==0]=1
    thres[thres==255]=0
    thres[thres==1]=255
    x,y,w,h = cv2.boundingRect(thres)
    cv2.rectangle(thres, (x, y+50), (x+w, y+h), (255, 0, 0), 1)
    # roi_r = image.crop((2020,320,2420,720))

    # plt.set_cmap('viridis')
    # j = 0
    # for i in range(400):
    #     if thres[i,y-1] == 255:
    #         print(j)
    #         break
    #     else: 
    #         j += 1
    # roi_l2 = thres[x:j, y-1:290]
    # roi_l3 = thres[j:x+h, y-1:290]
    rec = thres[y:y+h,x:x+w]
    contours2,hierarchy2 = cv2.findContours(rec,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    print(cv2.contourArea(contours2[0])/(w*h))
    plt.imshow(thres)

    roi_r = image[0:400, 530:800]
    ret_r, thres_r = cv2.threshold(roi_r,100,255,cv2.THRESH_BINARY)
    contours_r1,hierarchy_r1 = cv2.findContours(thres_r,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    print(len(contours_r1))
    area_r = []
    for i in range(len(contours_r1)):
        area_r.append(cv2.contourArea(contours_r1[i]))
    max_idx_r = np.argmax(area_r)
    for k in range(len(contours_r1)):
        if k != max_idx_r:
            cv2.fillPoly(thres_r, [contours_r1[k]], 255)
    
    thres_r[thres_r==0]=1
    thres_r[thres_r==255]=0
    thres_r[thres_r==1]=255
    x_r,y_r,w_r,h_r = cv2.boundingRect(thres_r)

    rec_r = thres_r[y_r:y_r+h_r,x_r:x_r+w_r]
    # cv2.rectangle(thres_r, (x_r, y_r), (x_r+w_r, y_r+h_r), (255, 0, 0), 2)
    contours_r2,hierarchy_r2 = cv2.findContours(rec_r,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    print(cv2.contourArea(contours_r2[0])/(w_r*h_r))
    plt.subplot(1, 2, 2)
    plt.imshow(rec_r)
# plt.show()    
    
    
```
